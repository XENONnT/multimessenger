How to discretize colormaps;

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
cmap = matplotlib.cm.get_cmap('jet')

x = np.linspace(0, 2*np.pi, 200)
y = np.cos(x)

for i in range(50):
    plt.plot(x,y+i/2, c=cmap(i/50))
```
